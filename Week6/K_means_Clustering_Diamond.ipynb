```python
import pandas as pd

# Gantilah 'nama_file.csv' dengan nama file dataset yang Anda inginkan
file_path = '/content/drive/MyDrive/Dataset/diamonds.csv'

# Membaca dataset dari file CSV
dataset = pd.read_csv(file_path)

# Mengubah dataset menjadi dataframe
df = pd.DataFrame(dataset)
df.head()
```

```python
# Menampilkan seluruh nama kolom
print(df.columns)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming 'df' is your DataFrame with one-hot encoded categorical variables and other attributes
attributes = df[['Unnamed: 0', 'carat', 'depth', 'table', 'price', 'x', 'y', 'z']]

# Inisialisasi list untuk menyimpan nilai inertia
inertia = []

# Menjalankan K-Means dengan jumlah kluster dari 1 hingga 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(attributes)
    inertia.append(kmeans.inertia_)

# Membuat plot elbow method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='-', color='b')
plt.xlabel('Jumlah Kluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Kolom yang akan digunakan untuk clustering
features = df[['Unnamed: 0', 'carat', 'depth', 'table', 'price', 'x', 'y', 'z']]

# Membuat model K-Means dengan 5 kluster
kmeans = KMeans(n_clusters=5, random_state=0)
df['label'] = kmeans.fit_predict(features)

# Visualisasi berdasarkan label kluster
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b', 'c', 'm']
for label, color in zip(range(5), colors):
    cluster_data = df[df['label'] == label]
    plt.scatter(cluster_data['carat'], cluster_data['price'], c=color, label=f'Cluster {label}')

plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 4], s=200, c='k', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
import numpy as np

# Daftar untuk menyimpan nilai Calinski-Harabasz Score
calinski_scores = []

# Mencoba jumlah kluster (k) dari 2 hingga 10
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(features)
    calinski_avg = calinski_harabasz_score(features, labels)
    calinski_scores.append(calinski_avg)

# Memplot Calinski-Harabasz Score
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), calinski_scores, marker='o', linestyle='-', color='b')
plt.title('Calinski-Harabasz Score for K-Means Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Calinski-Harabasz Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()

# Menampilkan nilai Calinski-Harabasz Score
for k, score in zip(range(2, 11), calinski_scores):
    print(f"Calinski-Harabasz Score for k={k}: {score:.4f}")
```
